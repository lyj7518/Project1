# 데이터 통합

```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
```

```python
# 데이터 불러오기
def csv_r(name):
    file_name = "/content/drive/MyDrive/🐹전종설🐹/코드/유진티비/" + name + ".csv"
    df = pd.read_csv(file_name, encoding='utf-8', index_col=0)
    return df
```

```python
df_rain = pd.read_csv("/content/drive/MyDrive/🐹전종설🐹/코드/유진티비/일별 서울 강수량 데이터(2015~2022).csv", encoding="CP949", index_col=0)
```

```python
df_coal = pd.read_csv("/content/drive/MyDrive/🐹전종설🐹/코드/유진티비/주별 유연탄 가격(2015-2022).csv", encoding="utf-8", index_col=0)
df_coal = df_coal.drop(df_coal.columns[[1, 2, 3, 4]], axis=1)
```

```python
df_smp = csv_r("smp(2015-2022)")
df_gas = csv_r("일별 천연가스 가격 데이터(2015~2022)")
df_kospi = csv_r("일별 코스피지수 데이터 (2015-2022)")
df_ExRate = csv_r("일별 환율 데이터(2015-2022)")
```
